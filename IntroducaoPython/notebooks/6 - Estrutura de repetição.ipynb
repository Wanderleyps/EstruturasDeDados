# Estruturas de repetição - for

```
print(1)
print(2)
print(3)
print(4)
print(5)
```

```
for numero in range(1,7):
  print(numero)
```

```
#range(__start: SupportsIndex, __stop: SupportsIndex, __step: SupportsIndex)
for numero in range(10, -1, -1):
  print(numero)
```

```
5 + 4 + 3 + 2 + 1
```

```
soma = 0
for numero in range(1, 6):
  soma = soma + numero
  #print(soma)

print(soma)
```

```
#Percorrer texto
palavra = 'sorvete'
for letra in palavra:
  #print(letra)
  if letra == 'v':
    print('Achou a letra v')
```

```
#for aninhados
for i in range(0,5):
  print(i)
  print('---')
  for j in range(0,3):
    print(j)
  print()
```

# Estruturas de repetição - while

```
numero = 1
while numero < 6:
  print(numero)
  numero = numero + 1
print('---')
```

```
numero = 5
while numero > 0:
  print(numero)
  numero -= 1
```

```
1 + 2 + 3 + 4 + 5
```

```
soma = 0
numero = 1
while numero < 6:
  soma += numero
  numero += 1
print(soma)
```

```
numero = 0
while numero < 1 or numero > 10:
  numero = int(input('Digite um número de 1 a 10: '))
```

Python não possui uma estrutura de controle específica chamada "do-while". No entanto, é possível alcançar o mesmo efeito usando uma estrutura de loop while em combinação com uma condição que verifica no final do loop se deve continuar executando

```
while True:
    # Código a ser executado dentro do loop

    resposta = input("Deseja continuar? (s/n): ")
    if resposta.lower() != 's':
        break
```
