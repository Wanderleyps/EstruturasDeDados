Construções de seleção, assim como em qualquer construção de controle, podem ser **aninhadas** de forma que o resultado de uma pergunta ajuda a decidir o que perguntar em seguida.

```
nota = int(input('Digite a nota: '))
if nota >= 90:
   print('A')
else:
   if nota >=80:
      print('B')
   else:
      if nota >= 70:
         print('C')
      else:
         if nota >= 60:
            print('D')
         else:
            print('F')
```

Uma sintaxe alternativa para esse tipo de seleção aninhada usa o palavra-chave **elif**. O else e o próximo if são combinados de modo a eliminar a necessidade de níveis adicionais de aninhamento.

```
nota = int(input('Digite a nota: '))
if nota >= 90:
   print('A')
elif nota >=80:
   print('B')
elif nota >= 70:
   print('C')
elif nota >= 60:
   print('D')
else:
   print('F') #note que o else final ainda é necessário para fornecer o caso default
```

```
x = 2
y = 4
if (x > 1) and (y % 2 == 0):
  print('x é maior que 1 e y é par')
else:
  print('Uma ou nenhuma das condições foram satisfeitas')
```
