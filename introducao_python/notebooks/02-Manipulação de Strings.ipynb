# Manipulação de strings

```
a = 'casaco'
print(a)
```

```
maiuscula = a.upper()
a = a.upper()
print(a)
#print(b)
print(maiuscula)
```

```
minuscula = maiuscula.lower()
print(minuscula)
```

```
# Primeira letra em maiúsculo
#a = 'wanderley pereira'
capital = a.capitalize()
print(capital)
```

Selecionar partes específicas de uma string

```
# Pegar da posição 0 até a posição 4 da string
a = 'casaco'
metade_palavra = a[0:4]
print(metade_palavra)
```

```
# Pega da terceira posição até o final
ultimas_letras = a[3:]
print(ultimas_letras)
```

Trocar caracteres com replace

```
b = a.replace('aco', 'inha')
print(a)
print(b)
```

```
c = a.replace('o', 'a')
print(c)
```

```
#Saber o tamanho de uma string
e = ' casaco '
print(len(e))
```
